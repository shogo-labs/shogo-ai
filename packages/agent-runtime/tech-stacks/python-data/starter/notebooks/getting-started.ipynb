# Getting Started

Welcome to your data science workspace! This notebook demonstrates common patterns for data analysis and visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
np.random.seed(42)
dates = pd.date_range('2024-01-01', periods=90, freq='D')
df = pd.DataFrame({
    'date': dates,
    'revenue': np.random.normal(1000, 200, 90).cumsum(),
    'users': np.random.poisson(50, 90),
    'category': np.random.choice(['A', 'B', 'C'], 90)
})

print(f'Dataset: {len(df)} rows')
df.head()
```

```python
# Quick visualization
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(df['date'], df['revenue'])
axes[0].set_title('Cumulative Revenue')
axes[0].set_xlabel('Date')

df.groupby('category')['users'].mean().plot(kind='bar', ax=axes[1])
axes[1].set_title('Average Users by Category')

plt.tight_layout()
plt.savefig('../output/figures/overview.png', dpi=150)
plt.show()
```
